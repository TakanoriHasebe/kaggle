### Statics

```python
data.head()
```

```python
# 置き換え
data['class'] = data['class'].map({'Normal' : 0, 'Abnormal' : 1}).astype(int)
data.rename(columns={"class" : "Abnormal"}, inplace=True)
data.rename(columns={"pelvic_tilt numeric" : "pelvic_tilt_numeric"}, inplace=True)
```

```python
data.head()
```

```python
# 複数の説明変数を持つモデルを推定する
lm_model = smf.ols("Abnormal ~  pelvic_incidence + pelvic_tilt_numeric + lumbar_lordosis_angle + sacral_slope + pelvic_radius + degree_spondylolisthesis", data = data).fit()
```

```python
# Type II ANOVAの結果
sm.stats.anova_lm(lm_model, typ=2).round(3)
```

```python
# AICの計算
lm_model.aic.round(3)
```

### Machine Learng by Statical Point of view

```python
# 教師データの作成
y = data['Abnormal']
```

```python
# 学習データの作成
x = data[['pelvic_tilt_numeric', 'pelvic_tilt_numeric', 'lumbar_lordosis_angle', 'degree_spondylolisthesis', 'sacral_slope', 'pelvic_radius']]
```

```python
y = y.values
```

```python
x = x.values
```

```python
(x_train, x_test,
 y_train, y_test) = train_test_split(
    x, y, test_size=0.3, random_state=0,
)
```

### Decision Tree

```python
dt = DecisionTreeClassifier()
dt.fit(x_train, y_train)
```

```python
dt.score(x_test, y_test)
```

### Random Forest

```python
rf = RandomForestClassifier(n_estimators=10)
rf.fit(x_train, y_train)
```

```python
rf.score(x_test, y_test)
```

### Voting Classifier

```python
lr = LogisticRegression()
dt = DecisionTreeClassifier()
svm = SVC(kernel='poly', degree=2)
rf = RandomForestClassifier(n_estimators=10)
```

```python
evc = VotingClassifier(estimators = [('lr', lr), ('svm', svm), ('rf', rf)], voting='hard')
```

```python
evc.fit(x_train, y_train)
```

```python
evc.score(x_test, y_test)
```

### Bagging

```python
bg = BaggingClassifier(DecisionTreeClassifier(), max_samples=0.5, max_features=1.0, n_estimators=20)
bg.fit(x_train, y_train)
```

```python
bg.score(x_test, y_test)
```

### Boosting

```python
abd = AdaBoostClassifier(DecisionTreeClassifier(), n_estimators=10, learning_rate=1)
abd.fit(x_train, y_train)
```

```python
abd.score(x_test, y_test)
```

### K-NEAREST NEIGHBORS (KNN)

```python
# KNN
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 3)
x, y = data.loc[:, data.columns != 'class'], data.loc[:, "class"]
knn.fit(x, y)
prediction = knn.predict(x)
print('Prediction: {}'.format(prediction))
```

```python
# train_test_split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3,random_state = 1)
knn = KNeighborsClassifier(n_neighbors = 3)
x,y = data.loc[:,data.columns != 'class'], data.loc[:,'class']
knn.fit(x_train,y_train)
prediction = knn.predict(x_test)
#print('Prediction: {}'.format(prediction))
print('With KNN (K=3) accuracy is: ',knn.score(x_test,y_test)) # accuracy
```

Random ForestやVoting ClassifierよりもK-NNの方が精度が高いことがわかる。<br>

### Kの値によっての違いをみる

```python
# Model complexity
neig = np.arange(1, 25)
train_accuracy = []
test_accuracy = []
# Loop over different values of k
for i, k in enumerate (neig):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(x_train,y_train)
    train_accuracy.append(knn.score(x_train, y_train))
    test_accuracy.append(knn.score(x_test, y_test))
    
plt.figure(figsize=[13,8])
plt.plot(neig, test_accuracy, label = 'Testing Accuracy')
plt.plot(neig, train_accuracy, label = 'Training Accuracy')
plt.legend()
plt.title('-value VS Accuracy')
plt.xlabel('Number of Neighbors')
plt.ylabel('Accuracy')
plt.xticks(neig)
# plt.savefig('graph.png')
plt.show()
print("Best accuracy is {} with K = {}".format(np.max(test_accuracy),1+test_accuracy.index(np.max(test_accuracy))))
```
